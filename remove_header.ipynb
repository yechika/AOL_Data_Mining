# Filter Kolom Dataset Banjir

Memfilter dataset untuk menyisakan hanya kolom yang dibutuhkan untuk analisis

```python
import pandas as pd
import numpy as np

print("Libraries imported successfully!")
```

## 1. Load Data

```python
df = pd.read_csv('data_banjir_joined_clean.csv')

print("Dataset loaded successfully!")
print("=" * 60)
print(f"Shape: {df.shape}")
print(f"Columns: {len(df.columns)}")
print("\nAll columns in dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:3d}. {col}")
```

## 2. Filter Kolom yang Dibutuhkan

```python
selected_columns = [
    'kabupaten_kota',
    'kecamatan',
    'avg_rainfall',
    'max_rainfall',
    'avg_temperature',
    'elevation',
    'landcover_class',
    'ndvi',
    'slope',
    'soil_moisture',
    'year',
    'month',
    'banjir',
    'lat',
    'long',
    'jumlah_penduduk'
]

print("Kolom yang akan disimpan:")
print("=" * 60)
for i, col in enumerate(selected_columns, 1):
    print(f"{i:3d}. {col}")

print(f"\nTotal: {len(selected_columns)} kolom")
```

```python
print("Validasi kolom:")
print("=" * 60)

missing_columns = []
available_columns = []

for col in selected_columns:
    if col in df.columns:
        available_columns.append(col)
        print(f"‚úì {col}")
        print(f"OK {col}")
        missing_columns.append(col)
        print(f"‚úó {col} - TIDAK DITEMUKAN")
        print(f"NO {col} - TIDAK DITEMUKAN")
if missing_columns:
    print(f"\n‚ö†Ô∏è  WARNING: {len(missing_columns)} kolom tidak ditemukan:")
    print(f"\nWARNING: {len(missing_columns)} kolom tidak ditemukan:")
        print(f"  - {col}")
else:
    print(f"\n‚úÖ Semua {len(selected_columns)} kolom ditemukan!")
```

```python
df_filtered = df[available_columns].copy()

print("Dataset setelah filtering:")
print("=" * 60)
print(f"Shape sebelum: {df.shape}")
print(f"Shape sesudah: {df_filtered.shape}")
print(f"Kolom dikurangi: {df.shape[1]} ‚Üí {df_filtered.shape[1]} ({df.shape[1] - df_filtered.shape[1]} kolom dihapus)")

print("\nSample data (5 baris pertama):")
print(df_filtered.head())
```

## 2.5 Hapus Data yang Tidak Ter-join (Tanpa jumlah_penduduk)

```python
print("Statistik sebelum pembersihan:")
print("=" * 60)
print(f"Total baris: {len(df_filtered):,}")
print(f"Baris dengan jumlah_penduduk: {df_filtered['jumlah_penduduk'].notna().sum():,}")
print(f"Baris tanpa jumlah_penduduk: {df_filtered['jumlah_penduduk'].isna().sum():,}")
print(f"Persentase missing: {(df_filtered['jumlah_penduduk'].isna().sum() / len(df_filtered) * 100):.2f}%")

df_filtered_clean = df_filtered[df_filtered['jumlah_penduduk'].notna()].copy()

print("\n" + "=" * 60)
print("Statistik setelah pembersihan:")
print("=" * 60)
print(f"Total baris: {len(df_filtered_clean):,}")
print(f"Baris dihapus: {len(df_filtered) - len(df_filtered_clean):,}")
print(f"Baris tersisa: {(len(df_filtered_clean) / len(df_filtered) * 100):.2f}%")

print("\n‚úÖ Data yang tidak ter-join berhasil dihapus!")
print("\nData yang tidak ter-join berhasil dihapus!")
```

## 3. Cek Data Quality (Setelah Pembersihan)

```python
print("Missing Values:")
print("=" * 60)
missing_stats = df_filtered_clean.isnull().sum()
missing_percent = (missing_stats / len(df_filtered_clean) * 100).round(2)

for col in df_filtered_clean.columns:
    missing_count = missing_stats[col]
    missing_pct = missing_percent[col]
    if missing_count > 0:
        print(f"‚ö†Ô∏è  {col:25s}: {missing_count:6d} ({missing_pct:5.2f}%)")
        print(f"WARNING {col:25s}: {missing_count:6d} ({missing_pct:5.2f}%)")
        print(f"‚úì  {col:25s}: {missing_count:6d} ({missing_pct:5.2f}%)")
        print(f"OK     {col:25s}: {missing_count:6d} ({missing_pct:5.2f}%)")
print(f"\nTotal missing values: {df_filtered_clean.isnull().sum().sum():,}")
```

```python
print("Dataset Info:")
print("=" * 60)
df_filtered_clean.info()

print("\n" + "=" * 60)
print("Summary Statistics:")
print("=" * 60)
print(df_filtered_clean.describe())
```

## 4. Export ke File CSV Baru

```python
output_filename = 'data_banjir_filtered.csv'

try:
    df_filtered_clean.to_csv(output_filename, index=False, encoding='utf-8-sig')
    
    print("DATA BERHASIL DISIMPAN!")
    print("=" * 60)
    print(f"File: {output_filename}")
    print(f"Total rows: {len(df_filtered_clean):,}")
    print(f"Total columns: {len(df_filtered_clean.columns)}")
    
    import os
    file_size = os.path.getsize(output_filename) / (1024 * 1024)
    print(f"File size: {file_size:.2f} MB")
    
    original_size = os.path.getsize('data_banjir_joined_clean.csv') / (1024 * 1024)
    reduction = ((original_size - file_size) / original_size * 100)
    print(f"\nPengurangan ukuran: {original_size:.2f} MB -> {file_size:.2f} MB ({reduction:.1f}% lebih kecil)")
    
    print("\nKOLOM YANG DISIMPAN:")
    for i, col in enumerate(df_filtered_clean.columns, 1):
        print(f"  {i:2d}. {col}")
    
    print(f"\nData tidak ter-join (tanpa jumlah_penduduk) telah dihapus")
    print(f"Baris awal: {len(df):,} -> Baris akhir: {len(df_filtered_clean):,}")
    
except Exception as e:
    print(f"Error saat menyimpan file: {e}")
```
