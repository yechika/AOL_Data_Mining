```python
import geopandas as gpd

filename = 'kecamatan.geojson' 
df = gpd.read_file(filename)
print(type(df))
df

```

```python
print("Daftar kolom dalam DataFrame:")
print(df.columns.tolist())
print(f"\nTotal kolom: {len(df.columns)}")
```

```python
kolom_relevan = [
    'objectid', 'kode_kec_spatial', 'nama_prop', 'nama_kab', 'nama_kec',
    'jumlah_penduduk', 'kepadatan_penduduk', 'jumlah_kk', 'luas_wilayah',
    'pertumbuhan_2022', 'pertumbuhan_2021', 'pertumbuhan_2020',
    'perpindahan_pddk',
    'jumlah_desa', 'jumlah_kelurahan',
    'perdagangan', 'nelayan', 'wiraswasta',
    'tidak_blm_sekolah', 'slta', 's1'
]

df_filtered = df[kolom_relevan].copy()

print(f"DataFrame original: {len(df.columns)} kolom")
print(f"DataFrame filtered: {len(df_filtered.columns)} kolom")
print(f"Total baris: {len(df_filtered)}")
print("\nKolom yang dipilih:")
for i, col in enumerate(kolom_relevan, 1):
    print(f"  {i}. {col}")
    
df_filtered.head()
```

```python
output_filtered = 'kecamatan_filtered.csv'
df_filtered.to_csv(output_filtered, index=False, encoding='utf-8')
print(f'Data filtered berhasil diekspor ke: {output_filtered}')
```
